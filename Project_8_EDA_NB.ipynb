```python
import pandas as pd
```

```python
file = "data/Electric_Car_Sales_by_Model_in_USA.csv"
df = pd.read_csv(file)
df.head()
```

```python
df.describe()
```

```python
##Take only first 3 months of 2012
df_short = df[['Make', 'Model', 'janv-12', 'Feb 2012', 'mars-12']]
df_short.head()
```

```python
##Rename the years columns
df_short.rename(columns={"janv-12": "Jan 2012", "mars-12": "Mar 2012"}, inplace=True)
```

```python
df_short.head()
```

```python
df_short['Make'].unique()
```

```python
## How many models per Make there are?
model_count_by_make = df_short.groupby(['Make'])['Model'].count()
model_count_by_make
```

```python
##Which column has the most missing data? Sort the columns by sum of null values
df_short.isnull().sum().sort_values(ascending=False)
```

```python
df_cleaned = df_short.dropna(subset=['Jan 2012', 'Feb 2012', 'Mar 2012'])
```

```python
df_cleaned.head(10)
```

```python
## Quick chart to show number of electric vehicles sold by month
df_cleaned[['Jan 2012', 'Feb 2012', 'Mar 2012']].apply(pd.Series.sum).head().plot(kind="bar", color='blue')
```
